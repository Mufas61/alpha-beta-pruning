## Run the default algorithm

Build the graph.

```python
DEAPTH, BRANCHING = 3, 2
```

```python
nums = build_random_leafs(DEAPTH, BRANCHING)
order, graph = build_graph(nums, BRANCHING)
print_pretty(graph, order)
```

And run the alpha-beta-pruning with the visualization.

```python
desc = alpha_beta(graph, root=order[0])

viz = build_viz(order, graph, desc, branching=BRANCHING)

viz.render(filename='img/graph')
display(Image(filename='img/graph.png'))
```

## The same graph with ordering

```python
sorted_graph, sorted_order  = min_max_sort(graph, root=order[0])
print_pretty(sorted_graph, sorted_order)
```
